# Matplotlib

## Introduction

Matplotlib is the "grandfather" library of data visualization with Python. It was created by John Hunter. He created it to try to replicate MatLab's (another programming language).

It is an excellent 2D and 3D graphics library for generating scientific figures. 

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

Official Matplotlib web page: http://matplotlib.org/
    
## Importing

Import the `matplotlib.pyplot` module under the name `plt` (the tidy way):

```python
import matplotlib.pyplot as plt #used to visualize(pictorial representation) the data 
import numpy as np              #used to deals with numerical(number(int,float)) data
```

## Basic Matplotlib Commands

```python
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
variable_x=[1,2,3,4]
variable_y=[2,4,6,8]
```

```python
plot1 = plt.plot(variable_x,variable_y)    # Plot function will take atleast 2 variable to return the graph.
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.title('graph')
# plt.show()                    #show() function will remove the out[] line
```

```python
num1 = [1,2,3,4,5]
num2 = np.random.randint(1,10,5)   #This function will return 5 random integer between 1 to 10(excluded)
```

```python
plt.plot(num1,num2)
plt.xlabel("x_value")
plt.ylabel("y_value")
plt.title("Graph - x vs y")
plt.show()
```

## Creating Multiplots on Same Canvas

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--') # More on color options later
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-');
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.5, 0.3]) # left, bottom, width, height (range 0 to 1)
```

```python
# Creates blank canvas
fig = plt.figure()

# Plot on that set of axes
x = np.linspace(0, 2 * np.pi, 200)
y = np.linspace(0, 3 * np.pi, 200)

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.6, 0.3, 0.2]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

## Marker

```python
# plot-->'cml'--color, mark, connecting line
```

```python
plt.plot(num1,num2, 'y*-.')        # Here we used marker-->cml (Color-yellow, mark-star, connecting_line-dash_dot)
plt.xlabel("x_value")
plt.ylabel("y_value")
plt.title("Graph - x vs y")
```

```python
# Limit the x_axis value and y_axis value by using "xlim" & "ylim"
# xticks
```

```python
x_axis_value = [1,2,3,4,5]
y_axis_value = [20,23,25,28,30]
```

```python
# using x_limit--> xlim
plt.plot(x_axis_value,y_axis_value, 'ko--')     # k is used for black color
plt.xlim((2,4))                                 # Noticed? X_axis_value 1 & 5 is out of graph with their correspond y_axis_value
plt.xlabel("x_value")
plt.ylabel("y_value")
plt.title("Graph - x vs y")
```

```python
# using y_limit--> ylim
plt.plot(x_axis_value,y_axis_value, 'k^-')
plt.ylim((15,28))              #Here you will find y_axis_value 30 is out of graph with their correspond x_axis_value
plt.xlabel("x_value")
plt.ylabel("y_value")
plt.title("Graph - x vs y")
```

```python
# using xlim and ylim
plt.plot(x_axis_value,y_axis_value, 'k^-')
plt.xlim((2,5))
plt.ylim((15,30))
plt.xlabel("x_value")
plt.ylabel("y_value")
plt.title("Graph - x vs y")
```

```python
# xtick and ytick are used to change the tick of x_axis & y_axis
```

```python
# using xtick and ytick
plt.plot(x_axis_value,y_axis_value, 'k^-')
plt.xticks(x_axis_value,['a','b','c','d','e'])
plt.yticks(y_axis_value,(range(10,35,5)))
plt.xlabel("x_value")
plt.ylabel("y_value")
plt.title("Graph - x vs y")
```

```python
# drawing 2 graphs in a single plot

x_axis1=[1,2,3,4,5]
y_axis1 = np.random.randint(20,40,5)
y_axis2 = np.random.randint(20,40,5)
plt.plot(x_axis1,y_axis1,'ro:',label='xy1') # label_func is used to differentiate the graph 
plt.plot(x_axis1,y_axis2,'y*-',label='xy2')
plt.legend()   #A legend is an area describing the elements of the graph
```

```python
plt.figure(figsize=(10,5))
x=[1,2,3,4,5]
y1 = np.random.randint(20,40,5)
y2 = np.random.randint(20,40,5)
plt.xlim((2,4))
plt.ylim((5,30))
plt.xticks(x,['a','b','c','d','e'])
plt.yticks(np.arange(5,40,2.5))
plt.xlabel("x_value")
plt.ylabel("y_value")
plt.plot(x,y1,'ro:',label='xy1')
plt.plot(x,y2,'y*-',label='xy2')
plt.legend()
```

# Bar graph

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# types of choclates liked by people
# 1. dairy milk-45
# 2. snickers - 80
# 3. five star - 35
# 4. mars - 60
# 5. perk - 32
# 6. munch - 68
```

```python
choclates = ['Dairy_milk','Snickers','Five_star','Mars','Perk','Munch']
people = [45,80,35,60,32,68]
plt.bar(choclates,people,color=['k','r','g','b','c','m','y'])
```

```python
choclates = ['dairy milk','snickers','five star','mars','perk','munch']
people = [45,80,35,60,32,68]
plt.barh(choclates,people,color=['k','r','g','b','c','m','y'])        #barh-->used for horizontal graph
```

```python
height = np.random.randint(100,190,60)
```

```python
height
```

```python
plt.figure(figsize=(5,15))
x_axis = list(range(1,61))
plt.barh(x_axis,height)
```

```python
plt.figure(figsize=(5,15))
x_axis = list(range(1,61))
plt.yticks(np.arange(1,61))
plt.ylim(1,20)
plt.barh(x_axis,height)
```

# stack plot

```python
personname = ['abc','xyz','pqr']
day1 = [4,40,4]
day2 = [4,4,4]
day3 = [4,10,4]
plt.stackplot(personname,day1, day2,day3)
plt.show()
```

```python
activity = ['sleep','read','run']
day1 = [8,2,2]
day2 = [7,4,1]
day3 = [6,2,3]
day4 = [6,8,2]
plt.stackplot(activity,day1,day2,day3,day4)
plt.xlabel('Activity')
plt.ylabel('days')
plt.show()
```

# Pie chart

```python
# x=[h1,h2,h3,h4] not required
y=[5,12,8,10]
plt.pie(y,autopct='%.2f') # %->percentage,2f->2 decimal
plt.show()
```

```python
x=[5,3,2]
plt.pie(x,autopct='%.2f',explode=[0.1,0,0.1],shadow=True)
plt.show()
```

```python
x=[5,12,8,10]
plt.pie(x,autopct='%.2f%%',explode=[0.1,0,0.1,0,],shadow=True)
plt.show()
```
